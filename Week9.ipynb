```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np
from skimage import io
from skimage import transform
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.morphology import (erosion, dilation, closing, opening, area_closing, area_opening)
from skimage.color import rgb2gray
```

```python
image = cv.imread('/content/drive/MyDrive/PCVK/Images/Images/j.png')

kernel = np.ones((5,5), np.uint8)
dilation = cv.dilate(image, kernel, iterations = 1)

plt.subplot(121),plt.imshow(image),plt.title('Initial Image')
plt.subplot(122),plt.imshow(dilation),plt.title('Dilation Image')
```

```python
image1 = cv.imread('/content/drive/MyDrive/PCVK/Images/Images/j.png', 0)
m,n= image1.shape 
plt.imshow(image1, cmap="gray")

k=5
SE= np.ones((k,k), dtype=np.uint8)
constant= (k-1)//2

imageErosion = np.zeros((m,n), dtype=np.uint8)

for i in range(constant, m-constant):
  for j in range(constant,n-constant):
    temp= image1[i-constant:i+constant+1, j-constant:j+constant+1]
    product= temp*SE
    imageErosion[i,j]= np.min(product)
plt.subplot(121), plt.imshow(image1, cmap ="gray"), plt.title('Initial Image')
plt.subplot(122), plt.imshow(imageErosion, cmap="gray"), plt.title('Erosion Image')
```

```python
image2 = cv.imread('/content/drive/MyDrive/PCVK/Images/Images/j.png')
kernel = np.ones((5,5), np.uint8)
dilation = cv.morphologyEx(image2, cv.MORPH_DILATE, kernel)

plt.subplot(121), plt.imshow(image2), plt.title('Initial Image')
plt.subplot(122), plt.imshow(dilation), plt.title('Erosion Image')
```
